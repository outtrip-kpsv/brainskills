```python
from urllib.parse import unquote
```

```python
r = requests.get('https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F:1990-%D0%B5_%D0%B3%D0%BE%D0%B4%D1%8B')
```

```python
dt = BeautifulSoup(r.text, 'html.parser')
```

```python
def page_nav(url, cls):
    tmp = BeautifulSoup(requests.get(url).text, 'html.parser')
    return [(i.get('title',0), 'https://ru.wikipedia.org' + unquote(i.get('href',0))) for i in tmp('div', class_ = cls)[0]('a')]
```

```python
for i in page_nav('https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F:1990-%D0%B5_%D0%B3%D0%BE%D0%B4%D1%8B', 'mw-category'):
    print(*i)
```

```python
for i in page_nav('https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F:1990-%D0%B5_%D0%B3%D0%BE%D0%B4%D1%8B', 'mw-category'):
    print('='*50)
    print(*i)
    print('='*50)

    for j in page_nav(i[1], 'mw-category-generated'):
        print('-',*j)
```

# Задание 4

Творческое задание: найти среднюю стоимость детских ботинок на сайте ebay (возьмите первые 10 страниц, чтобы не захламлять трафик=) )

```python
def get_int(n):
    if n.startswith('От'):
        n = n.replace(u'\xa0', u'')
        tmp = n.split(' до ')
        tmp[0] = float(tmp[0][3:-5].replace(',','.'))
        tmp[1] = float(tmp[1][:-5].replace(',','.'))
    else:
        tmp = n.replace(u'\xa0', u'')
        tmp = [float(tmp[:-5].replace(',','.'))]
    return tmp
```

```python
def pg_load(srch, n_rng):
    all_price = []
    for j in range(1, n_rng+1):
        data = {
            '_from':'R40',
            '_nkw':'+'.join(srch.split()),
            '_sacat':'0',
            'LH_TitleDesc':'0',
            '_osacat':'0',
            '_ipg':'192',
            '_pgn': j
        }
        
        r = requests.get('https://www.ebay.com/sch/i.html', params = data, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36'})
        dt = BeautifulSoup(r.text, 'html.parser')
        
        for i in dt('li', class_='s-item'):
            tmp = i('span', class_='s-item__price')
            if len(tmp) != 0:
                all_price.extend(get_int(tmp[0].text))
        print('Пройдено страниц: {}, собранно значений: {}'.format(j, len(all_price)))
    return all_price
```

```python
all_price = pg_load("baby boots", 10)
```
