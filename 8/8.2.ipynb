```python
t = np.linspace(0, 60, 10000)

x=4.4*(np.cos(t)+np.cos(1.1*t)/1.1)
y=4.4*(np.sin(t)-np.sin(1.1*t)/1.1) 
```

```python
plt.figure(facecolor='grey')
plt.axes().set_aspect(1)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title('График')
plt.xticks(np.linspace(-10, 10, 11))
plt.yticks(np.linspace(-10, 10, 11))
plt.grid(color='b', linewidth=0.3, linestyle = ':')
plt.plot(x, y, color = 'magenta', label = 'Graph')
plt.legend()
```

# Задание 2

Изобразите 4 графика (2 х 2) для функций sin, cos, tan, cot с помощью subplot. (Будьте внимательны, в некоторых точках функции могут не существовать).

Красиво оформите эти графики, сделайте все подписи, подберите размеры так, чтобы надписи разных графиков не перекрывались.

```python
x = np.linspace(-10, 10, 1000)
```

```python
import seaborn as sns
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(12,12),facecolor='grey')

plt.subplot(2, 2, 1)
plt.title('График sin(x)')
plt.plot(x, np.sin(x), label = 'sin(x)')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.ylim(-1.5, 1.5)

plt.subplot(2, 2, 2)
plt.title('График cos(x)')
plt.plot(x, np.cos(x), label = 'cos(x)')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.ylim(-1.5, 1.5)

plt.subplot(2, 2, 3)
plt.title('График tan(x)')
plt.ylim([-4, 4])
plt.xlim([-4, 4])
y = np.tan(x)
y[:-1][np.diff(y) < 0] = np.nan
plt.plot(x, y, label = 'tan(x)')
plt.legend(loc='lower left')
plt.xlabel('Ось x')
plt.ylabel('Ось y')

plt.subplot(2, 2, 4)
plt.title('График cot(x)')
y = 1 / np.tan(x)
y[:-1][np.diff(y) > 0] = np.nan
plt.ylim([-4, 4])
plt.xlim([-4, 4])
plt.plot(x, y, label = 'cot(x)')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
```
